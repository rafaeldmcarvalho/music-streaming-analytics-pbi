```python
import pandas as pd
#loading
df = pd.read_csv('../data/spotify_user_behavior_realistic_50000_rows.csv')

#ETL process ----

# 1. Convert signup_date to datetime
df['signup_date'] = pd.to_datetime(df['signup_date'])

# 2. Transform "Yes"/"No" columns to 1 (True) and 0 (False)
ad_columns = ['ad_interaction', 'ad_conversion_to_subscription']
for col in ad_columns:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

# 3. Calculate Tenure (Months as a customer)
reference_date = pd.to_datetime('2025-12-31')
df['tenure_months'] = ((reference_date - df['signup_date']).dt.days / 30.44).astype(int)
df['tenure_years'] = (df['tenure_months'] / 12).astype(int)


# 4. Outlier Verification and Handling
# Checking max values
## print("Max hours before:", df['avg_listening_hours_per_week'].max())

# Export ---

df.to_csv('../data/spotify_cleaned.csv', index=False)
df.head()
```
